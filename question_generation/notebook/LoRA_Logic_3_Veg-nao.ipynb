### Unique Answers Distribution

```python
# Unique answers distribution

def analyze_unique_combinations(df, column_name):

    df_filtered = df[df[column_name].notna() & (df[column_name] != '')]

    df_filtered['normalized'] = df_filtered[column_name].apply(lambda x: ','.join(sorted(x.split(','))))

    unique_combinations = Counter(df_filtered['normalized'])
    num_rows_with_values = len(df_filtered)
    num_unique_combinations = len(unique_combinations)

    return num_rows_with_values, num_unique_combinations, unique_combinations


num_values, num_combinations, combinations = analyze_unique_combinations(l3_ques_veg_nao, 'answers')

print(f"Number of rows with values: {num_values}")
print(f"Number of unique combinations: {num_combinations}")
print("Unique combinations and their counts:")
for combination, count in combinations.items():
    print(f"{combination}: {count}")
```

### Generate diverse answers - nao

Key rules to generate diverse answers:

1. **Count Questions**: If a question in the 'questions' column contains "how many", count the number of items in the 'answers' column and store the count as a word in the new 'ground_truth_ans' column.

2. **Existence Questions**: For questions containing "Are there", add 'yes' to 'ground_truth_ans' if the 'answers' column has a value; add 'no' if 'answers' is None.

3. **Universal Questions**: For questions containing "Do all", compare 'visual_features' minus 'blend_noise/list_noise_non_veg' set with 'answers'. If they match, add 'yes' to 'ground_truth_ans'; otherwise, add 'no'.

#### Function to generate diverse answers

```python
# Function to generate ground_truth_ans, For how many, exist, universal questions. 

list_noise_non_veg = set(['tomato', 'lemon', 'watermelon', 'grape','bottle','pan', 'plant', 'yoghurt', 
                           'candy', 'orange', 'wine', 'pear','knife', 'pot', 'lime', 'banana', 'milk', 
                           'cheese','chocolate', 'plum'])


def count_items(answer):
    if answer == 'None' or pd.isna(answer):
        return 'none'
    return len(answer.split(', '))

def exist_answer(answer):
    return 'no' if answer == 'None' or pd.isna(answer) else 'yes'

def universal_answer(visual_features, answer, noise_set):
    visual_features_set = set(visual_features.split(', ')) if visual_features else set()
    answer_set = set(answer.split(', ')) if answer else set()
    return 'yes' if visual_features_set - noise_set == answer_set else 'no'


def generate_diverse_answer(row):
    question = row['questions_nlp'].lower()
    if 'how many' in question:
        return count_items(row['answers'])
    elif 'are there' in question or 'is there' in question:
        return exist_answer(row['answers'])
    return universal_answer(row.get('visual_features'), row['answers'], list_noise_non_veg)


df_l3 = diverse_l3_veg_nao_visible.copy()
df_l3['ans_diverse'] = df_l3.apply(generate_diverse_answer, axis=1)
df_l3.to_csv("/home/jingying/git_paper_1/diverse_questions/lr_questions_rose/questions_l2_l3_Veg/diverse_logic3_veg_nao_visible_full.csv", index=False)
print(df_l3.shape)
```
