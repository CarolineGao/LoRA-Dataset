### Unique answers distribution

```python
# Unique answers distribution 

def analyze_unique_combinations(df, column_name):

    df_filtered = df[df[column_name].notna() & (df[column_name] != '')]

    df_filtered['normalized'] = df_filtered[column_name].apply(lambda x: ','.join(sorted(x.split(','))))

    unique_combinations = Counter(df_filtered['normalized'])
    num_rows_with_values = len(df_filtered)
    num_unique_combinations = len(unique_combinations)

    return num_rows_with_values, num_unique_combinations, unique_combinations


num_values, num_combinations, combinations = analyze_unique_combinations(logic_why_t1_selected, 'fa')

print(f"Number of rows with values: {num_values}")
print(f"Number of unique combinations: {num_combinations}")
print("Unique combinations and their counts:")
for combination, count in combinations.items():
    print(f"{combination}: {count}")
```

### Generate Visible_object for Blender - why_selected

```python
# Extract keyword
def extract_keyword(question):
    match = re.search(r'\bis (\w+)', question)
    return match.group(1) if match else None

# Function to randomly select 11 items
def random_select_11(items):
    item_list = items.split(', ')
    if len(item_list) > 11:
        return ', '.join(random.sample(item_list, 11)) 
    return ', '.join(item_list) 


# Function to split 'fa' values into two halves
def split_fa_values(row):
    fa_values = row['answers'].split(', ')
    keyword_ans = row['keyword_ans']
    
    if keyword_ans in fa_values:
        fa_values.remove(keyword_ans)   
        
    random.shuffle(fa_values)
    mid_point = (len(fa_values) + 1) // 2
    ground_truth_ans = ', '.join([keyword_ans] + fa_values[:mid_point])
    spatial_exclude_ans = ', '.join(fa_values[mid_point:])
    return ground_truth_ans, spatial_exclude_ans


# Define blend_middle and blend_noise sets
blend_middle = set(['bread', 'milk', 'yogurt', 'wine', 'chocolate', 'candy'])

blend_noise = set([
'lemon', 'orange', 'grape', 'banana',  
'cheese', 'onion', 'tomato', 'carrot', 'eggplant', 'pumpkin', 'mushroom',
'cucumber', 'pepper', 'garlic', 'watermelon', 'avocado', 'potato', 'plum', 
'lime', 'asparagus', 'broccoli', 'leek', 'capsicum', 
'zucchini','cabbage', 'springonion', 'parsley', 'pear', 'apple', 'meat', 'egg' ])


# Generate Blender Visible Data
def add_objects(row):
    fa_values = row['fa']
    if fa_values == 'None':
        fa_set = set()
    else:
        fa_set = set(fa_values.split(', '))
    
    ground_truth_ans_set = set(row['ground_truth_ans'].split(', '))
    spatial_exclude_ans_set = set(row['spatial_exclude_ans'].split(', '))
    
    # Remove 'None' from sets
    ground_truth_ans_set.discard('None')
    
    # For available noise objects
    available_noise = list(blend_noise - fa_set)
    
    
    # Ground_Truth: For Right_side_object
    random.shuffle(available_noise)
    right_side_object = list(ground_truth_ans_set) + available_noise[:max(0, 6 - len(ground_truth_ans_set))]
    right_side_object_str = ', '.join(right_side_object).lstrip(', ')

    # Noise_Truth: For Left_side_object
    random.shuffle(available_noise)
    left_side_object = list(spatial_exclude_ans_set) + available_noise[:max(0, 6 - len(spatial_exclude_ans_set))]
    left_side_object_str = ', '.join(left_side_object).lstrip(', ')
    
    # For middle_object
    middle_object = random.sample(blend_middle, 1)[0]

    return right_side_object_str, left_side_object_str, middle_object

def process_data(df, filename):
    
    full_output_path = f"{base_path}{filename}_visible_full.csv"
    blend_output_path = f"{base_path}{filename}_blend.csv"

    # Apply the functions
    df['keyword_ans'] = df['questions_nlp'].apply(extract_keyword)
    df['answers'] = df['fa'].apply(random_select_11)
    df[['ground_truth_ans', 'spatial_exclude_ans']] = df.apply(split_fa_values, axis=1, result_type='expand')
    df[['right_side_object', 'left_side_object', 'middle_object']] = df.apply(add_objects, axis=1, result_type='expand')
    blend_df = df[['right_side_object', 'left_side_object', 'middle_object']]

    df.to_csv(full_output_path)
    blend_df.to_csv(blend_output_path)

base_path = "/home/jingying/git_paper_1/diverse_questions/lr_questions_rose/questions_why/"

process_data(logic_why_t1_selected, "logic_why_t1_selected")
print(logic_why_t1_selected.shape)
```

```python
spatial_info_options_1 = [
    "on the left side of the table,",
    "on the left side of the middle object,",
    "in the front of the first and second cabinet, counting from left to right,",
    "in the front of the plants,",
    "on the dinosaur’s left side and to the left side of the middle object,",
    "between the dinosaur’s left side and the middle object,",
    "in the front of the condiments or on the left side of the middle object,",
    "on the right side of the five red round fruit but on the left side of the middle object,",
    "between the food which is the best source of lycopene and the middle object,",
    "on the same half side of the table as the green toy,",
    "on the same side as the dinosaur,",
    "on the half of the table opposite the leopard,",
    "on the left side if drawing a line between the plate of lime and the middle object,",
    "under sunlight or the ones on the edge of the table between cucumber and middle object,",
    "not on the same side as the cutting board next to the cabinet,",
    "not on the same half of the table as the orange pumpkin on the right corner,",
    "between the eggs, plants, and the middle object,"
]


spatial_info_options_2 = [
    "not under the sunlight,",
    "on the right side of the table,",
    "on the right side of the middle object,",
    "in the front of the third and fourth cabinet, counting from left to right,",
    "among the furthest 6 food items from the window,",
    "among the furthest 6 food items from the dinosaur,",
    "among the furthest 6 food items from the orange pumpkin next to the window,",
    "among the furthest 6 food items from the purple food next to the window,",
    "further from the window and not under the sunlight,",
    "on the opposite half side of the table as the two bottles,",
    "on the opposite half side of the table as the bowls,",
    "among the 6 furthest food items from the source of light in the image,"
]


def insert_sentence(original_question, sentence):
    key_phrases = ["want a fruit"]
    
    for phrase in key_phrases:
        if phrase in original_question.lower():
            parts = original_question.split(phrase, 1)
            return parts[0] + phrase + " " + sentence + parts[1]

    return original_question


# Method 1
def modify_questions(df):
    df['questions_spatial'] = df['questions_nlp']
    for i in range(len(df)):
        middle_object = df.at[i, 'middle_object']
        
        if i < 2000:  # For the first 2000 rows
            spatial_info_part_1 = [info.replace("middle object", middle_object) for info in spatial_info_options_1]
            sentence_to_add = random.choice(spatial_info_part_1)
        else:  # For rows from 2000 to the end
            spatial_info_part_2 = [info.replace("middle object", middle_object) for info in spatial_info_options_2]
            sentence_to_add = random.choice(spatial_info_part_2)
            
        df.at[i, 'questions_spatial'] = insert_sentence(df.at[i, 'questions_spatial'], sentence_to_add)
    
    return df


df = l3_why_t1_visible
l3_why_t1_visible_full = modify_questions(df)
l3_why_t1_visible_full.to_csv("/home/jingying/git_paper_1/diverse_questions/lr_questions_rose/questions_why/questions_why_t1/logic_why_t1_selected_visible_full.csv")
print(l3_why_t1_visible_full.shape)
```
